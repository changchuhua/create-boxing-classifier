# Background
In this project, we will make use of pre-trained, existing libraries for real-time keypoint detection for human body parts.<br/>
[Openpose](https://github.com/CMU-Perceptual-Computing-Lab/openpose) is a keypoint detection library released in approximately [2018](https://arxiv.org/pdf/1812.08008.pdf) written in C++.<br/>

In the journal released by the Openpose team, Openpose is described as a
>bottomup representation of association scores via Part Affinity
Fields (PAFs),<br/> a set of 2D vector fields that encode the
location and orientation of limbs over the image domain.

Simply described, Part Affinity Fields are vector fields containing information of specific limbs or body parts within an image, for a bottom up approach of matching the limbs to identify whole persons within the image. An image describing this process is shown below.
